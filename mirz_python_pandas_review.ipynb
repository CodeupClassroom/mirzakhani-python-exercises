# Filtering in Pandas

```python
from pydataset import data
```

```python
df = data('tips')
df
```

```python
df.tip
```

```python
df[df.tip > 5]
```

```python
df.sex == 'Male'
```

```python
df[df.sex == 'Male']
```

```python
tip_more_than_5 = df[df.tip > 5]
tip_more_than_5
```

```python
tip_more_than_5[tip_more_than_5.sex == 'Male']
```

python: if I want to say and i just say and

pandas: if I want to say and between columns I say &

```python
df[(df.tip > 5) & (df.sex == 'Male')]
```

### What about or? Answer: |

```python
df[(df.tip > 5) | (df.sex == 'Male')]
```

```python
def find_tip_and_sex(given_df, tip_amount, sex_of_guest):
    filtered_df = given_df[(given_df.tip > tip_amount) & (given_df.sex == sex_of_guest)]
    
    return filtered_df
```

```python
assert (find_tip_and_sex(df, 5, 'Male') == df[(df.tip > 5) & (df.sex == 'Male')]).all().all()
```

```python
find_tip_and_sex(df, 4, 'Female')
```
